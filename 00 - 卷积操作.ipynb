```python
import torch.nn as nn
import torch
from matplotlib import pyplot as plt
```

```python
""" nn.Conv2d
    - in_channels 输入的通道数
    - out_channels 输出的通道数
    - kernel_size 卷积核的尺寸
    - stride 步长，用来控制卷积核移动间隔
    - padding 输入边沿扩边操作
    - padding_mode 扩边的方式
    - bias 是否使用偏置
    - ...
"""
def plot_image(img):
    fig = plt.figure()
    for i in range(6):
        plt.subplot(2, 3, i + 1)
        plt.tight_layout()
        
        plt.imshow(img[i][0], cmap='gray', interpolation='none')
        plt.xticks([])
        plt.yticks([])
    plt.show()

image = torch.rand(6, 1, 28, 28)

print(image.shape)

plot_image(image)

c = nn.Conv2d(1, 3, 2, 2)

c_image = c(image)

print(c_image.shape)

plot_image(c_image.detach().numpy())
```
